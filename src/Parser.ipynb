```python
from Bio import SeqIO
from tqdm import tqdm
import numpy as np
```

```python
a=SeqIO.parse("GCA_000001405.15_GRCh38_genomic.fna/GCA_000001405.15_GRCh38_genomic.fna", "fasta")
```

```python
record=[[],[],[],[]]
for seq_record in tqdm(a):
    n=0
    if n <300:
        record[0].append(seq_record.id)
        record[1].append(seq_record.seq)
        record[2].append(len(seq_record))
        n+=1
    else: 
        print("Loaded!") 
        break
```

```python
def overlapping(s, ks):
    kmers = []
    for a in tqdm(range(len(s))):
        kmers.append(s[a:a+np.random.choice(ks)])
    return kmers
```

```python
for each in tqdm(record[1]):
    b=overlapping(each, [3,4,5])
    record[3].append(b)
    print ("Overlapping is done!")
```
